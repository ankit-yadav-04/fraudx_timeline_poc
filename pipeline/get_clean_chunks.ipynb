## Take doc X-Ray and get clean chunks

```python
import json
from pathlib import Path
from typing import Dict, List, Any

def transform_chunks(input_data: Dict[str, Any]) -> List[Dict[str, Any]]:
    """
    Keep only selected keys from each chunk and add sequential chunk_number.
    Output format: {"chunks": [ ... ]}
    """

    required_keys = ["boundingBoxes", "chunk", "pageNumbers", "suggestedText"]
    source_chunks = input_data.get("chunks", []) # this is the list of chunks

    total_chunks_encountered = len(source_chunks)
    invalid_chunks = 0
    transformed: List[Dict[str, Any]] = []

    for ch in source_chunks:
        if not isinstance(ch, dict):
            invalid_chunks += 1
            continue

        new_chunk = {k: ch.get(k) for k in required_keys}
        
        # Sequential numbering only for valid output chunks
        new_chunk["chunk_number"] = len(transformed) + 1
        transformed.append(new_chunk)

    print(
        f"Processing summary: encountered {total_chunks_encountered} chunks, "
        f"invalid {invalid_chunks}."
    )

    return {"chunks": transformed}

if __name__ == "__main__":
    
    input_file = Path("/home/ankit/smartsense_code/fraudx_timeline_poc/rough_jsons/15308/input3.json")
    output_file = Path("/home/ankit/smartsense_code/fraudx_timeline_poc/rough_jsons/15308/input3_pass0.json")

    with input_file.open("r", encoding="utf-8") as f:
        data = json.load(f)
    
    result = transform_chunks(data)

    with output_file.open("w", encoding="utf-8") as f:
        data = json.dump(result, f, indent=2, ensure_ascii=False)

    print(f"Done. Wrote {len(result['chunks'])} chunks to {output_file}")
    
```
