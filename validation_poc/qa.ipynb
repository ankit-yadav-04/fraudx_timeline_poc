```python
from groundx import Document, AsyncGroundX
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
client = AsyncGroundX(api_key=os.getenv("GROUNDX_API_KEY"))
```

```python
# query = "when did incident or this accident took place?"
query = "who is the patient"

bucket_id = 22997
generic_search = await client.search.content(
    id=bucket_id, query=query, n=10, verbosity=2
)

generic_results = generic_search.model_dump()

for i, result in enumerate(generic_results["search"]["results"]):
    print("Result of chunk", i + 1)
    # print("Document ID: ", result["document_id"])
    # print("Document Name: ", result["file_name"])
    print("Text: ", result["text"])
    print("Suggested text: ", result["suggested_text"])

    print("-" * 100)
```
